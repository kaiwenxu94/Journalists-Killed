```python
import pandas as pd
```

```python
df = pd.read_csv('journalistskilled.csv')
```

```python
year=[]
for i in range(0,1324):
    individualyear=df['date'][i][-4:]
    year.append(individualyear)
df['years']=year
df
```

```python
country=[]
for i in range(0,1324):
    individualcountry=df['location'][i].split()[-1]
    country.append(individualcountry)
country
df['country']=country
df
```

```python
dictallcountries=df['country'].value_counts().to_dict()
countryname=list(dictallcountries)
ctcode = pd.read_csv('country code.csv')
for word in countryname:
    for k in range(len(ctcode)):
        if word==ctcode['Country'][k]:
            n=countryname.index(word)
            countryname.remove(word)
            countryname.insert(n,ctcode['code'][k])
countryname
```

```python
dfc = pd.read_csv('media freedom index.csv')
dfc
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(color_codes=True)
fig,ax4 = plt.subplots()
ax4.scatter(dfc['Press Freedom Index in 2018'],dfc['Deaths Number of Journalists Till 2018'], color="r")
sns.regplot(dfc['Press Freedom Index in 2018'],dfc['Deaths Number of Journalists Till 2018'], color="r")
ax4.set_title('Correlation between Number of Jounalists killed and Press Freedom Index in 2018',fontsize=20)
fig.set_size_inches(14,7)
plt.xticks(size = 14)
plt.yticks(size = 14)
ax4.set_ylabel("Deaths Number of Journalists Till 2018", size = 15)
ax4.set_xlabel("Press Freedom Index in 2018", size = 15)
fig.savefig('Correlation between Number of Jounalists killed and Press Freedom Index in 2018.png')
```
