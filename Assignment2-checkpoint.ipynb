```python
import pandas as pd
df = pd.read_csv('journalistskilled.csv')
year=[]
for i in range(0,1324):
    individualyear=df['date'][i][-4:]
    year.append(individualyear)
df['years']=year

country=[]
for i in range(0,1324):
    individualcountry=df['location'][i].split()[-1]
    country.append(individualcountry)
country
df['country']=country
```

# 1. What are the most dangerous countries for journalists?

From the bar chart below, we can see that Iraq is the most dangerous state for journalists. From 1992 to 2018, 186 news workers were killed there. Then follows Syria where 123 reporters died.

```python
for a in [113,250,676,966]:
    df['country'][a]='Central African Republic'
for b in [609,631,632]:
    df['country'][b]='Dominican Republic'
dfcountry=df['country'].replace({'Syria':'Syrian Arab Republic','Russia':'Russian Federation','Territory':'Israel','Lanka':'Sri Lanka','Bosnia':'Bosnia and Herzegovina','Leone':'Sierra Leone','Libya':'Libyan Arab Jamahiriya','USA':'United States','Yugoslavia':'Serbia','Venezuela':'Bolivarian Republic of Venezuela','Iran':'Islamic Republic of Iran','Africa':'South Africa','Salvador':'El Salvador','Coast':'Cote d’Ivoire','Timor':'Timor-Leste','Bolivia':'Plurinational State of Bolivia','Arabia':'Saudi Arabia','Faso':'Burkina Faso','Rica':'Costa Rica','UK':'United Kingdom'}).value_counts()
dfcountryten=df['country'].replace({'Syria':'Syrian Arab Republic','Russia':'Russian Federation','Territory':'Israel','Lanka':'Sri Lanka','Bosnia':'Bosnia and Herzegovina','Leone':'Sierra Leone','Libya':'Libyan Arab Jamahiriya','USA':'United States','Yugoslavia':'Serbia','Venezuela':'Bolivarian Republic of Venezuela','Iran':'Islamic Republic of Iran','Africa':'South Africa','Salvador':'El Salvador','Coast':'Cote d’Ivoire','Timor':'Timor-Leste','Bolivia':'Plurinational State of Bolivia','Arabia':'Saudi Arabia','Faso':'Burkina Faso','Rica':'Costa Rica','UK':'United Kingdom'}).value_counts()[:10]
ax = dfcountryten.plot.barh()
ax.set_xlabel('Number of Journalists Killed')
ax.set_title('The Top 10 Most Dangerous Countries for Jounalists')
```

# 2. How many journalists were killed each year since 1992?

We can see that 2009 was the year when the most journalists were killed, while 2002 the least. Also, there were 2 journalists whose date of death still remained unclear.

```python
df['years'].value_counts()
dictyears=df['years'].replace({'nown':'Unknown','007)':'Unknown'}).value_counts().to_dict()
sortedlist=sorted(dictyears.items(), key = lambda k: k[0])
sorteddict=dict((x, y) for x, y in sortedlist)
pd.Series(sorteddict)
ax1=pd.Series(sorteddict).plot.bar()
ax1.set_title('Number of Journalists Killed from 1992-2018')
```

# 3. What did the reporters die of?

As is shown in the pie chart below, nearly 64.1% of news workers were killed by murder. 22.6% of them died under crossfire and 12.8% during dangerous assignment. The deaths of 2 journalists cannot be ascertained.

```python
dfdeathtype=df['type of death'].value_counts()
dfdeathtype
explode = (0, 0.1, 0, 0)
ax2=dfdeathtype.plot.pie(explode=explode,figsize=(6, 6),autopct='%1.1f%%',shadow=True, startangle=90)
ax2.set_title('Types of Death')
```

# 4. Were the journalists killed freelance or formal employees?

The pie chart shows most reporters were formal staff.

```python
df['organization'].value_counts()
dforganization=df['organization'].replace({'freelance':'Freelance','Freelancer':'Freelance'}).value_counts()
number_of_formal_staff=0
for i in range(1,len(dforganization)):
    number_of_formal_staff=number_of_formal_staff+dforganization[i]
freelancer_vs_staff = pd.Series([dforganization['Freelance'],number_of_formal_staff], index=['Number of Freelancers', 'Number of Formal Staff'])
freelancer_vs_staff
ax3=freelancer_vs_staff.plot.pie(figsize=(6, 6),autopct='%1.1f%%',shadow=True, startangle=90)
ax3.set_title('Number of Killed Freelancers vs Formal Staff')
```

For the dead news workers hired by news agencies, where were they from? The bar chart below shows that Reuters has the most reporters who got killed since 1992.

```python
dforg=dforganization.drop('Freelance')
ax3 = dforg[:20].plot.barh()
ax3.set_xlabel('Number of Journalists Killed')
ax3.set_title('The TOP20 Orgnizations that the Killed Journalists Belonged to')
```
