```python
import csv
import pandas as pd
from splinter import Browser
import time
df = pd.read_csv('journalistskilled.csv')
year=[]
for i in range(0,1324):
    individualyear=df['date'][i][-4:]
    year.append(individualyear)
df['years']=year

country=[]
for i in range(0,1324):
    individualcountry=df['location'][i].split()[-1]
    country.append(individualcountry)
country
df['country']=country
df
```

```python
def get_additional_info_of_each_journalist(browser):
    xjob =browser.find_by_id("jobs").text
    xmedium=browser.find_by_id("mediums").text
    xbeats_covered=browser.find_by_id("coverages").text
    xgender=browser.find_by_id("gender").text
    xlocal_or_foreign=browser.find_by_id("localOrForeign").text
    xfreelance=browser.find_by_id("freelance").text
    xtype_of_death=browser.find_by_id("typeOfDeath").text
    xsuspected_source_of_fire=browser.find_by_id("sourcesOfFire").text
    ximpunity=browser.find_by_id("impunity").text
    xtaken_captive=browser.find_by_id("captive").text
    xtortured=browser.find_by_id("tortured").text
    xthreatened=browser.find_by_id("threatened").text
    each_journalist=[i,xjob,xmedium,xbeats_covered,xgender,xlocal_or_foreign,xfreelance,xtype_of_death,xsuspected_source_of_fire,ximpunity,xtaken_captive,xtortured,xthreatened]
    return each_journalist
```

```python
results=[]
for i in range(662,993):
    try:
        url = df['url'][i]
        browser = Browser('chrome')
        browser.visit(url)
        time.sleep(1)
        every_journalist=get_additional_info_of_each_journalist(browser)
        results.append(every_journalist)
        quit()
    except Exception as e:
        print(e)
        print('Error on Journalist No %s' % i)
```

```python
with open('range(662,993).csv','w',newline='',encoding='utf8') as f:
    writer = csv.writer(f)
    header = ['number','job','medium','beats covered','gender','local or foreign','freelance','type of death','suspected source of fire','impunity','taken captive','tortured','threatened']
    writer.writerow(header)
    writer.writerows(results)
```

```python
with open('range(662,993).csv','r',encoding='utf8') as f:
	rows = csv.reader(f)
	for row in rows:
		print(row)
```
