```python
df_original= pd.read_csv('journalistskilled.csv')
year=[]
for i in range(0,1324):
    individualyear=df_original['date'][i][-4:]
    year.append(individualyear)
df_original['years']=year

country=[]
for i in range(0,1324):
    individualcountry=df_original['location'][i].split()[-1]
    country.append(individualcountry)
country
df_original['country']=country

df_original
```
